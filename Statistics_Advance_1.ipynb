**1) What is a random variable in probability theory?**

**Ans)** In probability theory, a **random variable** is a function that assigns a numerical value to each possible outcome of a random experiment. It serves as a bridge between abstract outcomes and numerical analysis, allowing us to quantify and analyze uncertainty. Random variables are categorized into two main types: **discrete**, which take on countable values (such as the number of heads in coin tosses), and **continuous**, which can take on any value within a range or interval (such as the exact amount of rainfall in a day). By representing outcomes with numbers, random variables make it possible to define probability distributions, calculate expectations, variances, and perform various statistical analyses. They are fundamental to the study and application of probability in real-world contexts.


**2) What are the types of random variables?**

**Ans)** Random variables can be classified into two main types: **discrete** and **continuous**, with a third, less common category known as **mixed** random variables. A **discrete random variable** is one that can take on a finite or countably infinite set of distinct values. These values are typically whole numbers, such as the number of heads in a series of coin tosses, the outcome of rolling a die, or the number of customers arriving at a store in an hour. In contrast, a **continuous random variable** can assume any value within a given range or interval on the real number line. These values are uncountably infinite and can include decimals or fractions, such as the exact height of a person, the amount of rainfall in a day, or the time it takes to complete a task. The **mixed random variable** combines elements of both discrete and continuous variables. For instance, it may have a discrete probability at a certain point (like the probability of an immediate machine failure) and a continuous distribution for other values (such as the duration it operates before failing). Each type plays a key role in probability theory and statistical modeling, depending on the nature of the data being analyzed.


**3) What is the difference between discrete and continuous distributions?**

**Ans)** The primary difference between discrete and continuous distributions lies in the type of values the random variables can assume and how probabilities are represented. A **discrete distribution** is associated with a **discrete random variable**, which takes on specific, separate values—typically whole numbers. Examples include the number of emails received in an hour or the outcome of rolling a die. In discrete distributions, probabilities are assigned to each individual value, and the total of all these probabilities is equal to 1. In contrast, a **continuous distribution** corresponds to a **continuous random variable**, which can take on any value within a given range or interval, such as a person's weight or the time it takes to complete a task. For continuous distributions, the probability of the variable taking on an exact value is zero; instead, probabilities are assigned over **intervals** using a **probability density function (PDF)**. The total area under the PDF curve equals 1. Essentially, while discrete distributions deal with countable outcomes, continuous distributions handle uncountably infinite possibilities within intervals.


**4) What are probability distribution functions (PDF)?**

**Ans)** A **Probability Distribution Function (PDF)** is a function that describes how probabilities are assigned to the possible values of a random variable. For **discrete random variables**, this function is called a **Probability Mass Function (PMF)**, which gives the probability that the variable takes on a specific value. For example, the PMF of rolling a fair six-sided die assigns a probability of 1/6 to each face. For **continuous random variables**, the PDF describes the likelihood that the variable falls within a certain range rather than taking any exact value, since the probability at a single point is zero. In this case, the probability of the variable lying between two values is found by calculating the area under the PDF curve over that interval. Key properties of PDFs include that they are always non-negative and that the total probability across all possible values sums or integrates to 1. Essentially, PDFs provide a complete description of how probabilities are distributed for different types of random variables.


**5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

**Ans)** The **Cumulative Distribution Function (CDF)** and the **Probability Distribution Function (PDF)** serve related but distinct roles in describing random variables. The PDF (or PMF for discrete variables) shows how the probabilities are distributed over individual values or intervals—essentially giving the likelihood of the random variable taking on specific values or falling within certain ranges. In contrast, the CDF gives the **cumulative probability** that the random variable is less than or equal to a particular value. In other words, the CDF at a point $x$ tells you the probability that the variable will be at most $x$. For discrete variables, the CDF is the sum of the probabilities up to $x$, while for continuous variables, it is the integral of the PDF from negative infinity to $x$. The CDF is always a non-decreasing function that ranges from 0 to 1, providing a complete picture of the distribution’s behavior in terms of accumulated probability, whereas the PDF represents the density or likelihood at each point.


**6) What is a discrete uniform distribution?**

**Ans)** A **discrete uniform distribution** is a probability distribution where a finite number of outcomes are **equally likely** to occur. In other words, if a random variable follows a discrete uniform distribution, each possible value has the **same probability**. For example, when rolling a fair six-sided die, the outcomes 1 through 6 are all equally likely, each with a probability of $\frac{1}{6}$. The discrete uniform distribution is often used to model situations where there is no preference or bias among a fixed set of distinct outcomes. Its probability mass function (PMF) assigns a probability of $\frac{1}{n}$ to each of the $n$ possible outcomes, where $n$ is the number of equally likely values.


**7) What are the key properties of a Bernoulli distribution?**

**Ans)** The **Bernoulli distribution** is one of the simplest discrete probability distributions and models a random experiment with exactly two possible outcomes: **success** (usually coded as 1) and **failure** (coded as 0). Here are its key properties:

1. **Two possible outcomes:** The random variable $X$ can take only two values—1 (success) and 0 (failure).

2. **Parameter $p$:** The distribution is characterized by a single parameter $p$, where $0 \leq p \leq 1$, representing the probability of success (i.e., $P(X=1) = p$).

3. **Probability mass function (PMF):**

   $$
   P(X = x) = p^x (1 - p)^{1 - x} \quad \text{for } x \in \{0,1\}
   $$

4. **Mean (expected value):** **bold text**

   $$
   E(X) = p
   $$

5. **Variance:**

   $$
   \text{Var}(X) = p(1 - p)
   $$

6. **Support:** The sample space is $\{0,1\}$.

In summary, the Bernoulli distribution models a single trial with two outcomes, where $p$ quantifies the likelihood of success, making it fundamental for modeling binary events.


**8) What is the binomial distribution, and how is it used in probability?**

**Ans)** The **binomial distribution** models the number of successes in a fixed number of independent trials of a binary experiment, where each trial has only two possible outcomes: success or failure. Each trial has the same probability of success, denoted by $p$, and the number of trials is $n$. The binomial distribution gives the probability of obtaining exactly $k$ successes out of $n$ trials.

Mathematically, the probability mass function (PMF) of the binomial distribution is:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

where $\binom{n}{k}$ is the number of ways to choose $k$ successes from $n$ trials.

The binomial distribution is widely used in probability and statistics for modeling situations like the number of heads in a series of coin tosses, the number of defective items in a batch, or the number of patients who recover after treatment. It helps quantify the likelihood of different counts of successes when the trials are independent and identically distributed.


**9) What is the Poisson distribution and where is it applied?**

**Ans)** The **Poisson distribution** is a discrete probability distribution that models the number of times an event occurs within a fixed interval of time or space, under the assumption that these events happen **independently** and at a **constant average rate**. It’s often used when counting the number of rare or random events over a continuous domain.

The probability mass function (PMF) of the Poisson distribution is:

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

where:

* $k$ is the number of occurrences (0, 1, 2, ...),
* $\lambda$ is the average number of occurrences in the interval,
* $e$ is Euler’s number (approximately 2.71828).

### Applications of the Poisson Distribution:

* Counting the number of phone calls received by a call center in an hour.
* Number of decay events from a radioactive source over a fixed period.
* Number of emails arriving in your inbox per minute.
* Number of accidents at a traffic intersection per month.

In short, the Poisson distribution is ideal for modeling counts of events that occur randomly and independently over time or space.


**10) What is a continuous uniform distribution?**

**Ans)** A **continuous uniform distribution** is a probability distribution where a continuous random variable is equally likely to take any value within a specified interval $[a, b]$. This means that the probability density is constant across the entire range, and there is no preference for any value within the interval.

The probability density function (PDF) of a continuous uniform distribution is given by:

$$
f(x) = \frac{1}{b - a} \quad \text{for } a \leq x \leq b
$$

and zero elsewhere.

Because the density is constant, the probability that the variable falls within any sub-interval of $[a, b]$ is proportional to the length of that sub-interval. This distribution is often used to model situations where all outcomes within a range are equally likely, such as selecting a random number between 0 and 1 or modeling measurement errors uniformly distributed within a tolerance range.


**11) What are the characteristics of a normal distribution?**

**Ans)** The **normal distribution**, also known as the **Gaussian distribution**, is one of the most important and widely used continuous probability distributions. Its key characteristics are:

1. **Bell-shaped curve:** The graph of a normal distribution is symmetric and shaped like a bell, centered around the mean.

2. **Symmetry:** The distribution is perfectly symmetric about its mean ($\mu$), meaning the left and right sides are mirror images.

3. **Mean, Median, and Mode:** All three are equal and located at the center of the distribution.

4. **Defined by two parameters:**

   * **Mean ($\mu$)** — determines the location or center of the distribution.
   * **Standard deviation ($\sigma$)** — measures the spread or dispersion; larger $\sigma$ means a wider curve.

5. **Asymptotic tails:** The tails of the curve approach the horizontal axis but never touch it, extending infinitely in both directions.

6. **68-95-99.7 Rule:** Approximately 68% of data falls within 1 standard deviation of the mean, 95% within 2, and 99.7% within 3.

7. **Mathematical form:** Its probability density function (PDF) is

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

The normal distribution is widely used because many natural phenomena tend to be normally distributed, especially when influenced by many small, independent effects.


**12) What is the standard normal distribution, and why is it important?**

**Ans)** The **standard normal distribution** is a special case of the normal distribution with a **mean of 0** and a **standard deviation of 1**. It is often denoted as $Z \sim N(0,1)$. Its probability density function (PDF) is:

$$
f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}
$$

Because it is standardized, the standard normal distribution provides a common reference that allows us to compare and analyze data from any normal distribution by converting values into **z-scores**. A z-score measures how many standard deviations a data point is from the mean, calculated by:

$$
z = \frac{x - \mu}{\sigma}
$$


### Why is it important?

1. **Simplifies calculations:** By converting any normal variable to the standard normal form, we can use standard tables (z-tables) or software functions to find probabilities and percentiles without recalculating for each distribution’s parameters.

2. **Foundation for inferential statistics:** Many statistical methods, hypothesis tests, and confidence intervals rely on the standard normal distribution.

3. **Universality:** Because of the central limit theorem, sums or averages of many independent random variables tend to follow a normal distribution, making the standard normal a cornerstone in probability and statistics.

In short, the standard normal distribution is a key tool for standardizing and understanding data modeled by normal distributions.


**13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

**Ans)**The **Central Limit Theorem (CLT)** is a fundamental concept in statistics that states that, regardless of the original distribution of a population, the distribution of the **sample means** will approach a **normal distribution** as the sample size becomes large enough. More specifically, when you take sufficiently large random samples (usually $n \geq 30$) from any population with a finite mean and variance, the distribution of the sample means will be approximately normal, with the same mean as the population and a standard deviation equal to the population standard deviation divided by the square root of the sample size.


### Why is the CLT critical in statistics?

1. **Justifies using normal distribution:** It allows statisticians to apply techniques and inferential methods based on the normal distribution even when the population distribution is not normal.

2. **Simplifies analysis:** Because many statistics (like the sample mean) have approximately normal distributions for large samples, it simplifies the calculation of probabilities and confidence intervals.

3. **Foundation for hypothesis testing:** Most hypothesis tests and confidence interval procedures rely on the CLT to approximate sampling distributions.

In essence, the CLT provides the theoretical foundation for making inferences about populations using sample data, making it one of the most important theorems in statistics.


**14) How does the Central Limit Theorem relate to the normal distribution?**

**Ans)** The **Central Limit Theorem (CLT)** directly explains why the **normal distribution** plays such a central role in statistics. It states that, regardless of the shape of the original population distribution, the distribution of the **sample means** tends to become approximately normal as the sample size increases. This means that even if the underlying data are not normally distributed—whether skewed, uniform, or otherwise—the averages of sufficiently large samples drawn from that population will follow a normal distribution.

Because of the CLT, the normal distribution serves as a universal model for the behavior of sample means, enabling statisticians to use normal-based methods for inference, such as confidence intervals and hypothesis testing. In summary, the CLT provides the theoretical link that justifies using the normal distribution to approximate the distribution of many statistics derived from samples, making it a cornerstone of statistical analysis.


**15) What is the application of Z statistics in hypothesis testing?**

**Ans)** Z statistics are commonly used in hypothesis testing when the population variance is known or when the sample size is large enough for the Central Limit Theorem to apply. They help test whether a sample mean significantly differs from a known population mean or whether a population proportion matches a specified value. By calculating a Z statistic, which measures how many standard errors the sample statistic is away from the hypothesized population parameter, researchers can compare this value to critical points on the standard normal distribution. This comparison helps decide whether to reject or fail to reject the null hypothesis. Because the Z-test relies on the normal distribution, it provides a straightforward and widely applicable method for making statistical inferences about population parameters based on sample data.


**16) How do you calculate a Z-score, and what does it represent?**

**Ans)** A **Z-score** is calculated by taking the difference between an individual data point and the mean of the data set, then dividing that difference by the standard deviation. Mathematically, it’s expressed as:

$$
Z = \frac{X - \mu}{\sigma}
$$

where $X$ is the value of the data point, $\mu$ is the mean of the population, and $\sigma$ is the population standard deviation.

The Z-score represents how many standard deviations a particular value is from the mean. A positive Z-score indicates the value is above the mean, while a negative Z-score indicates it is below the mean. Z-scores are useful because they standardize different data points from different distributions onto a common scale, allowing comparison between values from different data sets or distributions. They also help determine the probability of observing a value within a normal distribution and are fundamental in hypothesis testing and statistical inference.


**17) What are point estimates and interval estimates in statistics?**

**Ans)** In statistics, **point estimates** and **interval estimates** are two ways to infer population parameters based on sample data.

A **point estimate** is a single value calculated from the sample that serves as the best guess or estimate of an unknown population parameter. For example, the sample mean ($\bar{x}$) is a point estimate of the population mean ($\mu$), and the sample proportion ($\hat{p}$) estimates the population proportion. While point estimates are straightforward, they don’t provide any information about the uncertainty or variability associated with the estimate.

On the other hand, an **interval estimate** provides a range of values, called a confidence interval, within which the population parameter is likely to lie, along with a specified level of confidence (such as 95%). Interval estimates account for sampling variability and give a measure of precision, showing not just a single estimate but a range that reflects the uncertainty inherent in using sample data to infer about the population. This makes interval estimates generally more informative for decision-making than point estimates alone.


**18) What is the significance of confidence intervals in statistical analysis?**
**Ans)** Confidence intervals are crucial in statistical analysis because they provide a range of values within which the true population parameter is likely to fall, rather than just giving a single point estimate. This range, accompanied by a confidence level (such as 95%), reflects the uncertainty and variability inherent in sampling. By showing how precise or reliable an estimate is, confidence intervals help researchers understand the degree of uncertainty in their conclusions. They also allow for better decision-making by indicating the range of plausible values, rather than assuming a fixed value. In essence, confidence intervals offer a more informative and nuanced picture of the data, making statistical results more meaningful and trustworthy.


**19) What is the relationship between a Z-score and a confidence interval?**

**Ans)** The **Z-score** and **confidence interval** are closely related in statistics because the Z-score determines the critical values used to construct confidence intervals for population parameters when the data follow a normal distribution or the sample size is large.

A confidence interval estimates a range around a sample statistic (like the mean) that is likely to contain the true population parameter with a certain confidence level (e.g., 95%). To build this interval, you start with the sample estimate and add and subtract a margin of error. This margin of error is calculated by multiplying the **standard error** by a critical **Z-score** that corresponds to the desired confidence level. For example, for a 95% confidence interval, the Z-score is approximately 1.96, which captures the middle 95% of the standard normal distribution.

In summary, the Z-score defines how far from the sample estimate you need to go to create a confidence interval that reflects your chosen level of certainty about the population parameter.


**20) How are Z-scores used to compare different distributions?**

**Ans)** Z-scores are used to compare values from different distributions by standardizing those values onto a common scale. Because raw scores from different datasets or distributions might have different means and variances, it’s difficult to directly compare them. Calculating the Z-score for a value converts it into the number of standard deviations it lies above or below its distribution’s mean. This standardization removes the units and scales differences, allowing you to directly compare how unusual or typical values are across different distributions. For example, a Z-score of +2 in one dataset and a Z-score of +2 in another both indicate that the values are 2 standard deviations above their respective means, making them comparable despite differences in original scales or units.


**21) What are the assumptions for applying the Central Limit Theorem?**

**Ans)** The Central Limit Theorem (CLT) relies on a few key assumptions to ensure that the distribution of sample means approximates a normal distribution as the sample size grows:

1. **Independence:** The sampled observations should be independent of each other. This is typically ensured by random sampling or when sampling is done with replacement.

2. **Sample size:** The sample size $n$ should be sufficiently large. While there’s no strict cutoff, a common rule of thumb is $n \geq 30$. For populations that are already roughly normal, smaller samples may suffice.

3. **Finite variance:** The population from which samples are drawn should have a finite mean and finite variance. If the population variance is infinite, the CLT does not hold.

4. **Random sampling:** The samples must be randomly selected from the population to avoid bias.

When these conditions are met, the CLT guarantees that the sampling distribution of the sample mean will approach a normal distribution, regardless of the population’s original shape.

**22) What is the concept of expected value in a probability distribution?**

**Ans)** The **expected value** of a probability distribution is essentially the long-run average or mean value that you would anticipate if you repeated an experiment or random process many times. It represents the **weighted average** of all possible outcomes, where each outcome is weighted by its probability of occurring. For a discrete random variable, the expected value is calculated by multiplying each possible value by its probability and then summing all these products. For continuous random variables, it involves integrating the product of the variable and its probability density function over all possible values. The expected value provides a measure of the "center" or "balance point" of the distribution and is fundamental in decision-making, risk assessment, and many statistical analyses.


**23) How does a probability distribution relate to the expected outcome of a random variable?**

**Ans)** A probability distribution describes all the possible values a random variable can take and the likelihood (probability) of each value occurring. The **expected outcome**, or expected value, of a random variable is directly derived from this distribution—it represents the average or mean value you would expect if the random experiment were repeated many times. Essentially, the expected value is a summary measure of the distribution that weights each possible outcome by its probability. So, the probability distribution provides the full picture of all possible outcomes and their chances, while the expected value condenses this information into a single representative value indicating the long-term average result.

```python
#2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)


import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a = 1  # Lower bound (inclusive)
b = 6  # Upper bound (inclusive)
n = b - a + 1  # Number of possible outcomes

# Generate x values (possible outcomes)
x = np.arange(a, b + 1)

# Calculate probability mass function (PMF)
pmf = np.ones(n) / n

# Create the plot
plt.figure(figsize=(8, 6))
plt.stem(x, pmf, basefmt=" ", linefmt="b-", markerfmt="bo")
plt.title("Discrete Uniform Distribution PMF (a=1, b=6)")
plt.xlabel("x")
plt.ylabel("Probability")
plt.grid(True, alpha=0.3)
plt.xticks(x)
plt.ylim(0, max(pmf) * 1.2)

# Show the plot
plt.show()
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pmf(k, p):
    """
    Calculate the probability of k in a Bernoulli distribution with success probability p.

    Parameters:
    k (int): The outcome (0 or 1).
    p (float): Probability of success (0 <= p <= 1).

    Returns:
    float: Probability of observing k.
    """
    if k not in (0, 1):
        return 0
    return p**k * (1 - p)**(1 - k)

# Example usage:
p = 0.7  # probability of success
print("P(X=1):", bernoulli_pmf(1, p))
print("P(X=0):", bernoulli_pmf(0, p))
```

```python
#5 Create a Poisson distribution and visualize it using Python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# --- Configuration for the Poisson Distribution ---
# Lambda (λ): This is the average rate of events occurring in a fixed interval of time or space.
# For example, if λ = 3, it means on average 3 events happen in that interval.
lambda_param = 3.0

# Range of k values: k represents the number of occurrences of an event.
# We'll plot the probability for k from 0 up to a certain limit.
# The upper limit is chosen to ensure the probabilities become very small,
# covering most of the distribution's significant mass.
k_values = np.arange(0, 15) # From 0 to 14 occurrences

# --- Calculate Poisson Probabilities ---
# The probability mass function (PMF) of a Poisson distribution is given by:
# P(X=k) = (λ^k * e^(-λ)) / k!
# We use scipy.stats.poisson.pmf to calculate these probabilities efficiently.
probabilities = poisson.pmf(k_values, lambda_param)

# --- Create the Visualization ---
plt.figure(figsize=(10, 6)) # Set the figure size for better readability
plt.bar(k_values, probabilities, color='skyblue', edgecolor='black', alpha=0.7)

# --- Add Labels and Title ---
plt.title(f'Poisson Distribution (λ = {lambda_param})', fontsize=16)
plt.xlabel('Number of Occurrences (k)', fontsize=12)
plt.ylabel('Probability P(X=k)', fontsize=12)
plt.xticks(k_values) # Ensure all k values are shown on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a grid for easier reading of probabilities

# --- Add text annotations for probabilities (optional, but helpful) ---
for i, prob in enumerate(probabilities):
    plt.text(k_values[i], prob + 0.005, f'{prob:.3f}', ha='center', va='bottom', fontsize=9)

# --- Display the plot ---
plt.show()

print(f"Poisson Distribution Parameters:")
print(f"  Lambda (λ): {lambda_param}")
print(f"  K values (number of occurrences): {k_values}")
print(f"  Corresponding Probabilities: {probabilities.round(4)}")
```

```python
#6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt

# Define the range of values
values = np.arange(1, 11)

# Since it's a discrete uniform distribution, each value has equal probability
probability = 1 / len(values)

# Calculate the CDF by cumulative summation of the probabilities
cdf = np.cumsum([probability] * len(values))

# Plot the CDF as a step plot
plt.step(values, cdf, where='post')
plt.title('CDF of Discrete Uniform Distribution (1 to 10)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(values)
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

```python
#9 Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (list or np.ndarray): The input dataset.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores (line and histogram)
    plt.figure(figsize=(12, 5))

    # Line plot
    plt.subplot(1, 2, 1)
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.title('Z-scores Line Plot')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)

    # Histogram
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=20, color='green', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
data_sample = np.random.normal(50, 10, 100)  # Simulated dataset
z_scores = calculate_and_plot_zscores(data_sample)
```

```python
# Generate random variables and calculate their corresponding probabilities using the binomial distribution


import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10      # Number of trials
p = 0.5     # Probability of success

# Generate random binomial samples
sample_size = 1000
samples = np.random.binomial(n, p, sample_size)

# Count frequency of each outcome
unique_values, counts = np.unique(samples, return_counts=True)

# Calculate PMF for each unique value
pmf_values = binom.pmf(unique_values, n, p)

# Print outcomes and corresponding probabilities
print("Outcome\tFrequency\tPMF")
for val, freq, prob in zip(unique_values, counts, pmf_values):
    print(f"{val}\t{freq}\t\t{prob:.4f}")

# Plot histogram of simulated outcomes
plt.bar(unique_values, counts / sample_size, alpha=0.6, label='Simulated Frequencies', color='skyblue', edgecolor='black')

# Overlay the theoretical PMF
plt.plot(unique_values, pmf_values, 'ro--', label='Theoretical PMF')

plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n + 1))
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset


import numpy as np
from scipy.stats import norm

# Population parameters
mu_0 = 100       # Hypothesized population mean
sigma = 15       # Known population standard deviation
alpha = 0.05     # Significance level

# Generate a sample dataset (simulate data)
np.random.seed(42)  # For reproducibility
sample = np.random.normal(loc=105, scale=sigma, size=30)
sample_mean = np.mean(sample)
n = len(sample)

# Calculate Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print results
print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

# Sample dataset (e.g., test scores or measurements)
data = np.array([78, 85, 92, 88, 76, 95, 89, 84, 91, 87])

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(data)

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# Critical t-value for two-tailed test (since population std dev is unknown)
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print the result
print(f"Sample mean: {sample_mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats

# Step 1: Generate data from a normal distribution
np.random.seed(0)
mu = 100            # True population mean
sigma = 15          # True population standard deviation
n = 50              # Sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
confidence = 0.95
alpha = 1 - confidence

# Step 3: Calculate the t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 4: Compute the margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 5: Compute confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 6: Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0       # Mean
sigma = 1    # Standard deviation

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the PDF
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Plot the PDF
plt.plot(x, pdf, color='blue', label=f'N({mu}, {sigma}²)')
plt.title('Normal Distribution - Probability Density Function')
plt.xlabel('x')
plt.ylabel('PDF')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
lambda_ = 3  # average rate (mean number of events)

# Values to evaluate CDF at (0 to 10)
x = np.arange(0, 11)

# Calculate CDF values
cdf_values = poisson.cdf(x, mu=lambda_)

# Print CDF values
for val, cdf_val in zip(x, cdf_values):
    print(f"P(X ≤ {val}) = {cdf_val:.4f}")

# Plot the CDF
plt.step(x, cdf_values, where='post', color='blue')
plt.title('CDF of Poisson Distribution (λ=3)')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

```python
#  Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

# Generate two sample datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=100)   # Dataset 1 with std dev ~5
data2 = np.random.normal(loc=50, scale=10, size=100)  # Dataset 2 with std dev ~10

# Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Visualize using boxplots
plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'], patch_artist=True,
            boxprops=dict(facecolor='lightblue'), medianprops=dict(color='red'))

plt.title('Comparison of Standard Deviations')
plt.ylabel('Values')

# Annotate the plot with std dev values
plt.text(1, max(data1) + 1, f'Std Dev: {std1:.2f}', horizontalalignment='center')
plt.text(2, max(data2) + 1, f'Std Dev: {std2:.2f}', horizontalalignment='center')

plt.grid(True, axis='y')
plt.show()
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate sample data (e.g., normal distribution)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)

# Apply Z-score normalization
data_z = zscore(data)

# Plot histograms for comparison
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Original data
axs[0].hist(data, bins=30, color='skyblue', edgecolor='black')
axs[0].set_title('Original Data Distribution')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

# Normalized data
axs[1].hist(data_z, bins=30, color='lightgreen', edgecolor='black')
axs[1].set_title('Z-score Normalized Data Distribution')
axs[1].set_xlabel('Z-score')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
